### Задание 1

```python
df_movies = pd.read_csv('ratings.csv')
df_movies.head()
```

```python
def add_rating(row):
    if row <= 2.0:
        return 'низкий рейтинг'
    elif row <= 4.0:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
df_movies['class'] = df_movies.rating.apply(add_rating)
df_movies.head()
```

### Задание 2

```python
df_keywords = pd.read_csv('keywords.csv')
df_keywords.head()
```

```python
geo_data = { 'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def define_region(row):
    for key, items in geo_data.items():
        for item in items:
            if item in row:
                return key
    return 'undefined'
```

```python
df_keywords['region'] = df_keywords.keyword.apply(define_region)
```

### Задание 3

```python
movies_df = pd.read_csv('movies.csv')
```

```python
def production_year(row):
    for year in range(1950, 2011): 
        if str(year) in row:
            return year
    return 1900
```

```python
movies_df['year'] = movies_df.title.apply(production_year)
movies_df.head()
```

```python
movies_after_1950 = movies_df[movies_df.year != 1900][['movieId','year']]
movies_after_1950 
```

```python
df_ratings = pd.read_csv('ratings.csv')
full_df = pd.merge(movies_after_1950, df_ratings[['movieId', 'rating']], on = 'movieId')
```

```python
%%time
full_df.groupby('year').rating.mean().sort_values(ascending = False)
```
