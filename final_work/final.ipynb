```python
left_people = df[df.left == 1]['satisfaction_level']

t_stat, p_value = stats.ttest_1samp(left_people, df.satisfaction_level.mean())
print(t_stat, p_value)

# Нулевая гипотеза - среднее значение выборки равно среднему значению генеральной совокупности.
# сравниваю уровень удовлетворенности тех, кто покинул компанию с общим. предполагаю, что у покинувших он ниже

# отвергаем нулевую гипотезу о том, что между переменными нет связи, потому что наше p-value очень низкое
print('Количество покинувших всего: ', len(left_people))
```

если бутстраппинг сделать, то выходит тоже очень маленькое p-value и мы также можем отвергнуть нулевую гипотезу

```python
sat_means = []
for i in range(3000):
    sample_ = left_people.sample(frac = 1, replace = True).mean()
    sat_means.append(sample_)
    
    
sns.distplot(sat_means)    
```

```python
sat_means_total = []
for i in range(3000):
    sample_ = df.satisfaction_level.sample(frac = 1, replace = True).mean()
    sat_means_total.append(sample_)
    
    
sns.distplot(sat_means_total)    
```

```python
t_stat, p_value = stats.ttest_ind(sat_means, sat_means_total)
print(t_stat, p_value)
```

```python
left_people = df[df.left == 1]['average_montly_hours']

t_stat, p_value = stats.ttest_1samp(left_people, df.average_montly_hours.mean())
print(t_stat, p_value)

# Нулевая гипотеза - среднее значение выборки равно среднему значению генеральной совокупности. 
# сравниваю количество отработанных часов покинувших компанию со средним количеством часов. 
# предполагаю, что они больше работали

# отвергаем нулевую гипотезу о том, что между переменными нет связи, потому что наше p-value очень низкое
print('Количество покинувших всего: ', len(left_people))
```

если бутстраппинг сделать, то выходит тоже очень маленькое p-value и мы также можем отвергнуть нулевую гипотезу

```python
hours_means = []
for i in range(3000):
    sample_ = left_people.sample(frac = 1, replace = True).mean()
    hours_means.append(sample_)
    
hours_means_total = []
for i in range(3000):
    sample_ = df.average_montly_hours.sample(frac = 1, replace = True).mean()
    hours_means_total.append(sample_)
    
t_stat, p_value = stats.ttest_ind(hours_means, hours_means_total)
print(t_stat, p_value)    

sns.distplot(hours_means)    
sns.distplot(hours_means_total)    
```

Так вышло, что я уже частично проверила выше в пятом пункте - у них больше рабочих часов и больше неудовлетворенность. 

Средний уровень удовлетворенности у тех, кто покинул компанию, ниже и среднего по всей ген.совокупности, что было
показано выше, но также и ниже чем средний уровень продолжающих работать. То же самое можно сказать и про среднее количество часов в месяц:

```python
left_people = df[df.left == 1]['satisfaction_level']
stayed_people = df[df.left == 0]['satisfaction_level']


t_stat, p_value = stats.ttest_ind(left_people, stayed_people)
print(t_stat, p_value)

# Нулевая гипотеза - среднее значение выборки равно среднему значению продолжающих работать.
# сравниваю уровень удовлетворенности тех, кто покинул компанию с теми, кто остался. предполагаю, что у них он ниже

# отвергаем нулевую гипотезу о том, что между переменными нет связи, потому что наше p-value очень низкое

print('Средний уровень удовлетворенности покинувших: ', np.mean(left_people))
print('Средний уровень удовлетворенности оставшихся: ',  np.mean(stayed_people))
```

```python
left_people = df[df.left == 1]['average_montly_hours']
stayed_people = df[df.left == 0]['average_montly_hours']

t_stat, p_value = stats.ttest_ind(left_people, stayed_people)
print(t_stat, p_value)

# Нулевая гипотеза - среднее значение выборки равно среднему значению продолжающих работать. 
# сравниваю количество отработанных часов покинувших компанию с теми, кто остался. 
# предполагаю, что они больше работали и поэтому покинули

# отвергаем нулевую гипотезу о том, что между переменными нет связи, потому что наше p-value очень низкое
print('Среднее количество часов в месяц покинувших: ', np.mean(left_people))
print('Среднее количество часов в месяц оставшихся: ',  np.mean(stayed_people))
```

Однако гипотеза про то, что у тех, кто покинул компанию, было больше инцидентов на работе, не подтвердилась.
