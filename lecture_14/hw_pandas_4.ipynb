### Задание 1

```python
import pandas as pd
```

```python
df = pd.read_csv('ratings.csv')
df.head()
```

```python
rating_sorted = df.groupby('userId').rating.count().reset_index()
rating_sorted = rating_sorted[rating_sorted.rating > 100]
rating_sorted
```

```python
rating_max=df.groupby('userId').max().reset_index()[['userId', 'timestamp']]
rating_min=df.groupby('userId').min().reset_index()[['userId', 'timestamp']]
```

```python
rating_sorted = rating_sorted.merge(rating_max, on='userId', how = 'left').merge(rating_min, on = 'userId', how = 'left')[['userId', 'timestamp_x', 'timestamp_y']]
rating_sorted.rename(columns={'timestamp_x': 'timestamp_max', 'timestamp_y': 'timestamp_min'}, inplace=True)

rating_sorted
```

```python
def max_lifetime(row):
    return row['timestamp_max'] - row['timestamp_min'] 
```

```python
rating_sorted['lifetime'] = rating_sorted.apply(max_lifetime, axis = 1) 
rating_sorted
```

```python
average_lifetime = rating_sorted.lifetime.mean()
average_lifetime
```
